# 딥러닝을 위한 PyTorch 활용법
## Chapter 2. Neural Network
- y = x**2 + 3

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.init as init
import torch.optim as optim
from torch.autograd import Variable

%matplotlib inline
```

```python
data_num = 10000
epoch_num = 10000

noise = init.normal_(torch.Tensor(data_num, 1), std=1)
x = init.uniform_(torch.Tensor(data_num, 1), -10, 10)

y = x**2 + 3
y_noise = y + noise
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(x.numpy(), y_noise.numpy(), color='g')
plt.scatter(x.numpy(), y.numpy(), color='r')
plt.xlabel('x')
plt.ylabel('y')
```

```python
model = nn.Sequential(
            nn.Linear(1,6),
            nn.ReLU(),
            nn.Linear(6,10),
            nn.ReLU(),
            nn.Linear(10, 20),
            nn.ReLU(),
            nn.Linear(20,1))

output = model(Variable(x))

loss_func = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr = 0.0001)
```

```python
loss_arr = []

label = Variable(y_noise)
for i in range(epoch_num):
    optimizer.zero_grad()
    
    output = model(Variable(x))
    loss = loss_func(output, label)
    loss.backward()
    
    optimizer.step()
    
    if i % 20 == 0:
        print(loss)
    loss_arr.append(loss)
    
```
